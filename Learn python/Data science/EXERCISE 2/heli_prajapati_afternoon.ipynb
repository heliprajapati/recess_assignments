Writing a Pandas program to compare the elements of two Pandas series

```python
import pandas as pd
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])
# Compare the elements of the two series
print(series1 == series2)
# Output:
# array([False, False, True, False, True])
```

Writing a Pandas program to add, subtract, multiply and divide two Pandas series.  

```python
# Create two pandas series
ser1 = pd.Series([2, 4, 6, 8, 14])
ser2 = pd.Series([1, 3, 5, 7, 9])
# Add the two series
print(ser1 + ser2)
# Output:
# [3, 7, 11, 15, 23]
# Subtract the two series
print(ser1 - ser2)
# Output:
# [1, 1, 1, 1, 5]
# Multiply the two series
print(ser1 * ser2)
# Output:
# [2, 12, 30, 56, 126]
# Divide the two series
print(ser1 / ser2)
# Output:
# [2.0, 1.3333333333333333, 1.2, 1.1428571428571428, 1.5555555555555556]
```

Writing a Pandas program to convert a dictionary to a Pandas series. 

```python
dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}
series1 = pd.Series(dictionary1)
print(series1)
```

Writing a Pandas program to convert a given series to an array.

```python
series = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])
array = series.to_numpy()
print(array)
```
