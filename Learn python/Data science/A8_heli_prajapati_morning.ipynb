```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('results.csv')
df
```

```python
print(df.info())
```

```python
df.columns
```

```python
#check for null values
df.isnull().sum().sort_values(ascending=False)
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#statistical details 
df.describe().T
```

```python
# Drop rows with missing values
print(df.dropna)
```

```python
# Fill missing values
df = df.fillna(80)
df
```

```python
# Remove wrong data
df.loc[2,'TOTAL CANDIDATES'] = 165
df
```

```python
# Remove wrong formats
df = df.replace(to_replace=r'[^\d.]', value=np.nan)
df
```

```python
# Remove duplicates
df.drop_duplicates()
```

```python
#correlation
cm = df.corr()
print(cm)
```

```python
#Heatmap
plt.figure(figsize=(10, 8))
sn.heatmap(cm, cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Scatterplot
# Extract the relevant columns for the scatter plot
x = df['YEAR']
y = df['TOTAL CANDIDATES']

# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(x, y)
plt.xlabel('Year')
plt.ylabel('Total candidates')
plt.title('Scatter Plot')
plt.show()
```
